```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman colleration"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
books_missing = books[~(books["description"].isna()) &
    ~(books["num_pages"].isna()) &
    ~(books["average_rating"].isna()) &
    ~(books["published_year"].isna())
]
```

```python
books_missing
```

```python
books_missing["categories"].value_counts().sort_values(ascending=False)
```

```python
word_counts = books_missing["description"].str.split().str.len()


books_missing.loc[:, "words_in_description"] = word_counts
```

```python
book_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

```python
combined_titles = np.where(
    book_missing_25_words["subtitle"].isna(),
    book_missing_25_words["title"],
    book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)
)

book_missing_25_words.loc[:, "title_and_subtitle"] = combined_titles
```

```python
book_missing_25_words
```

```python
tagged_descriptions = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)

book_missing_25_words.loc[:, "tagged_description"] = tagged_descriptions
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words
        .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
        .to_csv("books_cleaned.csv", index=False)
)
```
